```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scores = npzfile = np.load("scores.npz")['arr_0']
#scores = [*scores1, *scores2]
len(scores)
```

```python
scores_window = []
avg_scores = []
for score in scores:
    scores_window.append(score)
    if (len(scores_window)>100):
        scores_window.pop(0)
        avg_scores.append(np.average(np.float32(scores_window)))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid(True)
plt.plot(np.arange(len(scores)), scores)
plt.plot(np.arange(100,len(scores)), avg_scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
print(avg_scores[-1])
```
